```python
apple["Daily returns"]= apple['Adj Close'].pct_change()
apple.head()
```

```python
apple.dropna(axis = 0, inplace = True)  #Youll notice that the first value in the Daily Returns column is NaN. Well drop this row
apple.head(20)
```

```python
apple['Trend'].hist(bins = 50, figsize = (10,5)) 
plt.xlabel('Daily returns')
plt.ylabel('Frequency')
plt.show()
#satistics
apple.Daily_returns.describe()
```

```python
apple['Trend']=apple['Open'].pct_change()
apple.head()
```

```python
def trend(x):
  if x > -0.015 and x < 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x < 0.04:
    return 'Slight Positive'
  elif x > -0.015 and x < -0.04:
    return 'Slight Negative'
  elif x > 0.04 and x < 0.06:
    return 'Positive'
  elif x > -0.04 and x < -0.06:
    return 'Negative'
  elif x > 0.06 and x < 0.07:
    return 'Among top gainers'
  elif x > -0.06 and x < -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x < -0.07:
    return 'Bear drop'
apple['Trend']= (np.zeros(apple['Daily returns'].count()))
apple['Trend']= apple['Daily returns'].apply(lambda x:trend(x))
apple
```

```python
m_dict=dict()
plt.figure(figsize=(10,8))

for Trend in set(amazon.Trend):
  m_dict[Trend]=amazon[amazon.Trend==Trend].Trend.count()
plt.pie(m_dict.values,labels=m_dict.keys())
plt.show()
```

```python
apple_pie_data = apple.groupby('Trend')
pie_label = sorted([i for i in apple.loc[:, 'Trend'].unique()])
plt.pie(apple_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 1)

plt.show()
```
